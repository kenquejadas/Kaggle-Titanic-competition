# Step 1: Import libraries, read data

```python
# import data science libraries
import pandas as pd
import numpy as np
# upload and display training data
train_data = pd.read_csv('train.csv')
print('Training Data:')
train_data.head()
```

```python
# upload and display testing data
test_data = pd.read_csv('test.csv')
print('Testing Data:')
test_data.head()
```

# Step 2: Prepare data

```python
# Change Sex variable to binary to work with the ML model
train_data.replace({'male': 1, 'female': 0}, inplace=True)


# Embarked attribute might be significant, split that single column into three, one for each port of embarkation
train_data.rename(columns={'Embarked':'Cherbourg'}, inplace=True)
train_data.replace([np.inf, -np.inf], np.nan, inplace=True)
train_data.dropna(subset=["Cherbourg"], how="all", inplace=True)

train_data['Queenstown'] = train_data.loc[:, 'Cherbourg']
train_data['Southampton'] = train_data.loc[:, 'Cherbourg']

train_data['Cherbourg'].replace({'C' : 1, 'Q' : 0, 'S': 0}, inplace=True)
train_data['Queenstown'].replace({'C' : 0, 'Q' : 1, 'S': 0}, inplace=True)
train_data['Southampton'].replace({'C' : 0, 'Q' : 0, 'S': 1}, inplace=True)

# Missing Age values, drop rows with NaN
train_data = train_data.dropna()

train_data.head()
```

```python
# Do the same for testing data

# Change Sex variable to binary to work with the ML model
test_data.replace({'male': 1, 'female': 0}, inplace=True)


# Embarked attribute might be significant, split that single column into three, one for each port of embarkation
test_data.rename(columns={'Embarked':'Cherbourg'}, inplace=True)
test_data.replace([np.inf, -np.inf], np.nan, inplace=True)
test_data.dropna(subset=["Cherbourg"], how="all", inplace=True)

test_data['Queenstown'] = test_data.loc[:, 'Cherbourg']
test_data['Southampton'] = test_data.loc[:, 'Cherbourg']

test_data['Cherbourg'].replace({'C' : 1, 'Q' : 0, 'S': 0}, inplace=True)
test_data['Queenstown'].replace({'C' : 0, 'Q' : 1, 'S': 0}, inplace=True)
test_data['Southampton'].replace({'C' : 0, 'Q' : 0, 'S': 1}, inplace=True)

# Missing Age values, drop rows with NaN
test_data = test_data.dropna()

test_data.head()
```

# Step 3: Train ML model & evaluate

```python
# View correlations and hypothesize
train_corr = train_data.corr().abs()[["Survived"]]
train_corr = train_corr.sort_values(['Survived'], ascending = False)
train_corr
```

```python
x_train = train_data[['PassengerId','Sex','Pclass','Fare','Age','Parch','SibSp','Cherbourg','Southampton','Queenstown']]
y_train = train_data[['Survived']]

x_test = test_data[['PassengerId','Sex','Pclass','Fare','Age','Parch','SibSp','Cherbourg','Southampton','Queenstown']]
# For test_data, the 'Survived' attribute does not exist because that is what we are trying to predict

# Normalize the data in each data frame so all value ranges are the same
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train = sc.fit_transform(x_train)
X_test = sc.transform(x_test)
```

```python
# Train Decision Tree Model

from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_train,y_train,test_size = 0.2, random_state=1)

# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
