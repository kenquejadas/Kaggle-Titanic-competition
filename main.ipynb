# Step 1: Import libraries, read data

```python
# import data science libraries
import pandas as pd
import numpy as np
# upload and display training data
train_data = pd.read_csv('train.csv')
print('Training Data:')
train_data.head()
```

```python
# upload and display testing data
test_data = pd.read_csv('test.csv')
print('Testing Data:')
test_data.head()
```

# Step 2: Prepare data

```python
# Change Sex variable to binary to work with the ML model
train_data.replace({'male': 1, 'female': 0}, inplace=True)


# Embarked attribute might be significant, split that single column into three, one for each port of embarkation
train_data.rename(columns={'Embarked':'Cherbourg'}, inplace=True)
train_data.replace([np.inf, -np.inf], np.nan, inplace=True)
train_data.dropna(subset=["Cherbourg"], how="all", inplace=True)

train_data['Queenstown'] = train_data.loc[:, 'Cherbourg']
train_data['Southampton'] = train_data.loc[:, 'Cherbourg']

train_data['Cherbourg'].replace({'C' : 1, 'Q' : 0, 'S': 0}, inplace=True)
train_data['Queenstown'].replace({'C' : 0, 'Q' : 1, 'S': 0}, inplace=True)
train_data['Southampton'].replace({'C' : 0, 'Q' : 0, 'S': 1}, inplace=True)

train_data.head()
```

```python
# Do the same for testing data

# Change Sex variable to binary to work with the ML model
test_data.replace({'male': 1, 'female': 0}, inplace=True)


# Embarked attribute might be significant, split that single column into three, one for each port of embarkation
test_data.rename(columns={'Embarked':'Cherbourg'}, inplace=True)
test_data.replace([np.inf, -np.inf], np.nan, inplace=True)
test_data.dropna(subset=["Cherbourg"], how="all", inplace=True)

test_data['Queenstown'] = test_data.loc[:, 'Cherbourg']
test_data['Southampton'] = test_data.loc[:, 'Cherbourg']

test_data['Cherbourg'].replace({'C' : 1, 'Q' : 0, 'S': 0}, inplace=True)
test_data['Queenstown'].replace({'C' : 0, 'Q' : 1, 'S': 0}, inplace=True)
test_data['Southampton'].replace({'C' : 0, 'Q' : 0, 'S': 1}, inplace=True)

test_data.head()
```
