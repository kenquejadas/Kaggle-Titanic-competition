## Load training data, use to train ML model

```python
# Load and display training data
train_csv = "/kaggle/input/titanic/train.csv"
df = pd.read_csv(train_csv)
df.head()
```

```python
# Change Sex variable to binary
df_one = pd.get_dummies(df["Sex"])
df_two = pd.concat((df_one, df), axis=1)
df_two = df_two.drop(["Sex"], axis=1)
df_two = df_two.drop(["male"], axis=1)
result = df_two.rename(columns={"female": "Gender"})
# Gender Binary: 0 = Male, 1 = Female
result
```

```python
# Train a decision tree
train = pd.DataFrame(
    np.random.rand(100, 5),
    columns=['PassengerId', 'Pclass', 'Sex','Age', 'Survived'])

X = train[['PassengerId','Pclass', 'Sex','Age']]
y = train[['Survived']]
```

```python
from sklearn import preprocessing
from sklearn import utils

#convert y values to categorical values
lab = preprocessing.LabelEncoder()
y_transformed = lab.fit_transform(y)

#view transformed values
# print(y_transformed)

# Load test data
test_csv = "/kaggle/input/titanic/test.csv"
X_test = pd.read_csv(test_csv)
# ValueError: could not convert string to float: 'Kelly, Mr. James'
# Solution: Drop Name column
# Also drop columns with missing values
cols_with_missing = [col for col in X_test.columns 
                                 if X_test[col].isnull().any()]
X_test = X_test.drop(['Name'] + cols_with_missing, axis = 1)
# ValueError: could not convert string to float: 'male'
# Solution: Change Sex variable to binary
X_test = pd.get_dummies(X["Sex"])
X_pred = pd.concat((X_test, X), axis=1)
X_pred = X_pred.drop(["Sex"], axis=1)
# X_pred = X_pred.drop(["male"], axis=1)
X_pred.head()
```

```python
# Decision tree predictive model
from sklearn import tree
model = tree.DecisionTreeClassifier().fit(X,y_transformed)
y_pred = model.predict(X_pred) # ValueError: Input contains NaN, infinity or a value too large for dtype('float32').
```
